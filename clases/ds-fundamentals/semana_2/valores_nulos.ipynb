```python
import pandas as pd
import numpy as np
import time
```

```python
path = '../data/kaggle/nba_data_processed.csv'
df = pd.read_csv(filepath_or_buffer=path, sep= ',', header=0)
print(df.shape)
df.head(4)
```

```python
qsna=df.shape[0]-df.isnull().sum(axis=0)
qna=df.isnull().sum(axis=0)
ppna=round(100*(df.isnull().sum(axis=0)/df.shape[0]),2)
aux= {'datos sin NAs en q': qsna, 'Na en q': qna ,'Na en %': ppna}
na=pd.DataFrame(data=aux)
na.sort_values(by='Na en %',ascending=False)
```

```python
df.loc[df.isnull()['Player']]
```

```python
df.dropna(how='all',inplace=True)
df.shape
```

```python
qsna=df.shape[0]-df.isnull().sum(axis=0)
qna=df.isnull().sum(axis=0)
ppna=round(100*(df.isnull().sum(axis=0)/df.shape[0]),2)
aux= {'datos sin NAs en q': qsna, 'Na en q': qna ,'Na en %': ppna}
na=pd.DataFrame(data=aux)
na.sort_values(by='Na en %',ascending=False)
```

```python
df.loc[df.isnull()['FT%']].head(10)
```

```python
df.describe()['FT%']
```

```python
valor_inputacion = df['FT%'].mean()
type(df['FT%'].fillna(valor_inputacion))
```

```python
df['FT%'].fillna(valor_inputacion,inplace = True)
```

```python
qsna=df.shape[0]-df.isnull().sum(axis=0)
qna=df.isnull().sum(axis=0)
ppna=round(100*(df.isnull().sum(axis=0)/df.shape[0]),2)
aux= {'datos sin NAs en q': qsna, 'Na en q': qna ,'Na en %': ppna}
na=pd.DataFrame(data=aux)
na.sort_values(by='Na en %',ascending=False)
```

```python
start_time = time.time()
nuevos_valores = df['FT%'].fillna(valor_inputacion) # genera columna con valores no nulos originales y valores nulos imputados
df.loc[:,'FT%'] = nuevos_valores # reemplaza los valores nulos
execution_time = time.time() - start_time
print(execution_time*1000)
```

```python
qsna=df.shape[0]-df.isnull().sum(axis=0)
qna=df.isnull().sum(axis=0)
ppna=round(100*(df.isnull().sum(axis=0)/df.shape[0]),2)
aux= {'datos sin NAs en q': qsna, 'Na en q': qna ,'Na en %': ppna}
na=pd.DataFrame(data=aux)
na.sort_values(by='Na en %',ascending=False)
```

```python
start_time = time.time()
valor_inputacion = {'FT%':df['FT%'].mean()} # genera un diccionario con el valor de imputacion
df.fillna(valor_inputacion,inplace = True) # reemplaza los valores nulos
execution_time = time.time() - start_time
print(execution_time*1000)
```
