# Ejercicio 1: Verificar la edad para entrar en una fiesta
Descripción: El programa debe solicitar al usuario su edad y, en función de la misma, mostrar un mensaje indicando si puede entrar a una fiesta. Las reglas son las siguientes:
* Si el usuario tiene menos de 18 años, se mostrará un mensaje diciendo que no puede entrar.
* Si el usuario tiene entre 18 y 25 años (inclusive), se le permitirá entrar con una bebida sin alcohol.
* Si el usuario tiene más de 25 años, se le permitirá entrar con una bebida alcohólica.
###Objetivo
*Repasar cómo tomar input del usuario.
*Usar condicionales para determinar la salida.
*Mostrar mensajes usando output.
###Tareas:

1. Tengo que generar un mecanismo por el cual pueda ingresar un valor
2. Asignar el valor ingresado a una variable y ser capaz de identificar cuál de las 3 condiciones se cumple
3. Entregar como output el valor correspondiente a la condición cumplida

```python
while True:
    try:
        edad = int(input("Ingresa tu edad: "))
        if edad < 0:
            print("No puedes ingresar números negativos")
        elif edad >= 0 and edad < 18:
            print("No puedes entrar.")
        elif edad > 17 and edad < 26:
            print("Adelante. Puedes entrar con una bebida sin alcohol.")
        elif edad > 25:
            print("Adelante. Puedes entrar con una bebida alcohólica.")
        break
    except ValueError as e:
        print(f"Error tipo: {e}")
```

# Ejercicio 2: Transformar frases
Crea un código que tome la frase inicial entregada y la corrija de tal forma que quede igual a la frase final entregada. Donde:
* frase inicial es "hola, soMos el mejor cursó de de Data-Science de Coding_Dojo. yahoo!!!"
* frase final es "hola, somos el mejor curso de Data Science de Coding Dojo"

```python
frase = "hola, soMos el mejor cursó de de Data-Science de Coding_Dojo. yahoo!!!"
frase_final = "hola, somos el mejor curso de Data Science de Coding Dojo"
tildes = {'á':'a', 'é':'e', 'í':'i', 'ó':'o', 'ú':'u'}
```

```python
import string
nueva_frase = []

# Iteramos las palabras
for word in frase.split(): # frase.split() convierte cada palabra en un elemento de una lista
    aux = ''

    # Iteramos letra por letra las palabras
    for letter in word:
        # Si encontramos un punto, detenemos el ciclo de letras
        if letter == '.':
            break
        # Reemplazamos las tildes en las letras exceptuando la coma
        elif letter == ',' or letter not in string.punctuation:
            if letter in tildes.keys():
                aux += tildes[letter] # Reemplazamos las letras con tilde
            else:
                aux += letter
        # Reemplazamos guiones o guiones bajos por espacios
        elif letter == '_' or letter == '-':
            # Ponemos en mayuscula las letras que le siguen a - y _
            if aux[-1] == '-' or aux[-1] == '_':
                aux += letter.upper()
            else:
                aux += ' '

    auxiliar = ''

    for i in range(len(aux)):
        # Mantenemos las mayúsculas donde deben estar
        if i == 0 or aux[i-1] == ' ':
            auxiliar += aux[i]
        else:
            auxiliar += aux[i].lower()


    # Añadimos la palabra a la lista nueva_frase
    if not nueva_frase or auxiliar != nueva_frase[-1]:
        nueva_frase.append(auxiliar)

    # Si encontramos un punto, detenemos el ciclo de letras
    if '.' in word:
        break

frase = " ".join(nueva_frase)

# imprimimos el resultado
print(frase)

# Corroboramos que sean iguales
print(frase == frase_final)
```

# Ejercicio 3 : Verificar la contraseña de un usuario
Descripción: El programa debe pedir al usuario un nombre y una contraseña. Después de eso, verificará si la contraseña ingresada cumple con ciertos criterios:

* La contraseña debe tener al menos 8 caracteres.
* Debe contener al menos una letra mayúscula.
* Debe contener al menos un número.
* Si la contraseña cumple con todos los criterios, el programa dirá que el registro fue exitoso; si no, indicará qué condición no se cumplió y pedirá al usuario que lo intente nuevamente.
### Objetivos:

* Repasar cómo definir y usar funciones.
* Manejar input y output de manera dinámica.
* Usar condicionales para verificar múltiples criterios.
* Manipular cadenas de texto (uso de métodos como .isupper(), .isdigit(), etc.).

```python
import string
def user_register():
    excepciones = {'.', '_', '-'}
    no_permitido = set(string.punctuation) - excepciones
    while True:
        user = input("Ingrese su cuenta de usuario: ")
        if len(user) > 0:
            for char in user:
                if char in no_permitido:
                    print(f"Solo son permitidos los símbolos {excepciones}")
                    return user_register()
            return password_register(user)
        else:
            print("Debes escribir un nombre para la cuenta de usuario")

def password_register(user):
    while True:
        isdigit_flag = False
        islower_flag = False
        isupper_flag = False
        password = input("Ingrese su contraseña: ")

        if len(password) > 7:
            for char in password:
                if char.isdigit():
                    isdigit_flag = True
                if char.isupper():
                    isupper_flag = True
                if char.islower():
                    islower_flag = True

            if  isdigit_flag and islower_flag and isupper_flag:
                print("¡¡¡Registro exitoso!!!")
                return f"Tu cuenta de usuario es: {user} y tu contraseña es: {password}"
            else:
                print("La contraseña debe tener al menos 8 caracteres, un número, una letra mayúscula y una letra minúscula")
        else:
            print("La contraseña debe tener al menos 8 caracteres")


print("¡Bienvenido al Sistema de Registro de Cuenta de Usuario!\n")

print(user_register())
```
