Ejercicio: Análisis de una Competencia de Kaggle
Contexto:
Tienes un conjunto de datos en formato JSON que contiene información sobre una competencia de Kaggle. Los datos incluyen el ID del participante, el número de intentos, el tiempo invertido en la competencia y las puntuaciones obtenidas en cada intento. Algunos participantes tienen datos incompletos o valores nulos, que deberás manejar antes de realizar el análisis.
Objetivos:
Cargar y explorar un archivo JSON.
Identificar y manejar datos nulos.
Calcular estadísticas descriptivas (promedios, máximos, mínimos, etc.) para el número de intentos, el tiempo invertido y las puntuaciones.
Aplicar condicionales y operadores matemáticos para evaluar el rendimiento de los participantes (e.g., calcular el promedio de puntuaciones por intento).
Guardar el conjunto de datos limpio en un nuevo archivo JSON.

```python
import io
import pandas as pd
from google.colab import files

# Subir el archivo
uploaded = files.upload()

# Cargar el archivo JSON en un DataFrame
for filename in uploaded.keys():
    data = pd.read_json(io.BytesIO(uploaded[filename]))

# Explorar los primeros registros
print(data.head())

# Identificar datos nulos
print(data.isnull().sum())

# Manejar datos nulos solo en columnas numéricas
numerical_columns = data.select_dtypes(include=['number']).columns
data[numerical_columns] = data[numerical_columns].fillna(data[numerical_columns].mean())

# Estadísticas descriptivas para número de intentos, tiempo invertido y puntuaciones
print(data.describe())

# Calcular el promedio de puntuaciones por intento
# Assuming 'puntuaciones' column contains lists of scores
# We first calculate the sum of scores and then divide by attempts
data['promedio_puntuacion'] = data['puntuaciones'].apply(lambda x: sum( [i for i in x if i is not None] ) if isinstance(x, list) else x) / data['intentos'].replace(0, 1)
#If 'puntuaciones' is not a list, it uses the value directly. This handles cases where 'puntuaciones' might already be a single value instead of a list.

# Mostrar resultados
# Changed 'ID_participante' to 'id_participante' to match the actual column name
print(data[['id_participante', 'promedio_puntuacion']])

# Guardar datos limpios en un nuevo archivo JSON
data.to_json('archivo_limpio.json', orient='records', lines=True)

# Descargar el archivo limpio
files.download('archivo_limpio.json')
```
